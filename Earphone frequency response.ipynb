## Motivation
Hi, I bought some low quality earbuds and to my surprise they sound hecka garbage. It's as if the sound came from a speaker stuffed in a tin can. To fix this, I can play a frequency sweep (aka chirp) through the output and benchmark the received amplitude with a nice and much more expensive headset. After receiving, filtering, and normalizing the respective frequency response, multiple frequency points can be selected to apply compensation through an equalizer application. In my case, I used PulseAudio Multiband EQ for Ubuntu, though any equalizer should work.

```python
import numpy as np
import pyaudio
import sounddevice as sd
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib notebook
```

```python
fs = 44100
sd.default.samplerate = fs
```

## Testing
Making sure sound can be played and recorded

```python
# Speaker Test
duration = 5
sin_freq = 440
t = np.r_[:duration:1/fs]
myarray = np.sin(2*np.pi*sin_freq*t)
sd.play(myarray, fs)
```

```python
# Microphone Test
duration = 10
myrecording = sd.rec(int(duration * fs), samplerate=fs, channels=2)
```

```python
sd.play(myrecording, fs)
```

## Data logging
Obtaining samples involved pressing each earpiece against a cheap microphone I had laying around, and running the respective cell. The signal envelope (basically instantaneous magnitude, look up scipy.hilbert) is filtered with a 1000 tap box filter then normalized by max before plotting.

```python
f0 = 20
f1 = 20000
duration = 15
t = np.r_[:duration:1/fs]
chirp = np.sin(2*np.pi*(f0+(f1-f0)/duration * t/2)*t)
myrecording_earbud = sd.playrec(chirp, fs, channels=1).T[0]
# myrecording = sd.playrec(chirp, fs, channels=1).T[0]
```

```python
DIASO_env = abs(signal.hilbert(myrecording_earbud))
DIASO_filt = np.convolve(DIASO_env, np.ones(1000),'same')
DIASO_normalized = DIASO_filt / max(DIASO_filt)
plt.plot(f0+(f1-f0)/duration * t, DIASO_normalized)
plt.title("shitty diaso earbuds")
plt.xlabel("frequency (Hz)")
plt.ylabel("amplitude")
plt.show()
```

```python
AKG_env = abs(signal.hilbert(myrecording))
AKG_filt = np.convolve(AKG_env, np.ones(1000),'same')
AKG_normalized = AKG_filt / max(AKG_filt)
plt.plot(f0+(f1-f0)/duration * t,AKG_normalized)
plt.title("Nice AKG Headphones")
plt.xlabel("frequency (Hz)")
plt.ylabel("amplitude")
plt.show()
```

## Calculating Gains
Gains are a product, which are obtained by dividing the frequency response: desired / current. Gains are then converted to dB. My equalizer, like most, choose tunable frequencies in a log scale, so the x axis is plotted in log scale. For convenience I print the dB required to plug into the equalizer below, and these frequencies are paramaterized by changeable.

```python
changeable = np.array([50, 100, 156, 220, 311, 440, 622, 880, 1250, 1750, 2500, 3500, 5000, 10000, 20000])
changeable_indices = np.around(duration * (changeable - f0) / (f1 - f0) * fs).astype(int) - 1

gain = 20*np.log10(AKG_normalized/DIASO_normalized)
plt.plot(f0+(f1-f0)/duration * t,gain - np.mean(gain))
plt.scatter(changeable, (gain - np.mean(gain))[changeable_indices], c = 'g')
plt.title("gain")
plt.xlabel("frequency (Hz)")
plt.ylabel("dB")
plt.xscale('log')
plt.show()
```
